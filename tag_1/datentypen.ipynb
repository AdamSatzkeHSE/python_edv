### Primitive Datentypen, Variablen, Vergleichsoperatoren 

```python
# Strings / Zeichenketten - Datentypen zur Darstellung von Zeichen / Symbol in Textform.

print("Das ist eine Zeichenkette")
print("A")
print("17")
print(type("17"))

# Integers (Ganze Zahl), Float (Fließkommazahl), Complexe Zahlen
print(17)
print(type(17))

print(17.5)
print(type(17.5))
```

```python
# Datentypen sind je nach Operation wichtig / entscheidend.
# Datenkonvertierung in Python:

# Datenimport
print("17" + "18")

# Datentyp - Umwandlung auf gewünschtes Datentyp
value_1 = float("17")
value_2 = float("18")

# Datenaufarbeitung
result = value_1 + value_2

# Datenvisualisierung
# print("Ergebnis:", value_1 + value_2)
print("Ergebnis:", result, "Weitere Ergebnisse...")

# Variablen 

# print(17 + 18)
```

```python
# Weitere Datentypen: Bool
# print(True)
# print(False)

# Vergleichsoperatoren
print(3 > 10)
print(3 < 10)
print(3 == 3)
print(10 != 100)
print(100 >= 99)
print(90 <= 8)
```

```python
# (List, Tuple, Dict) -> Datenstrukturen
# Listen
import numpy as np

liste = [True, 23, 9.4, "Hallo Welt"] # Python erlaubt beliebige Datentypen in Listen zu speichern.
zahlen = [1, 2, 3]

result = []
for zahl in zahlen:
    result.append(zahl * 2)

# List-Methods / Listen-Methoden
result.pop()
result.append(2)
result.append(4)
print("Die 2 kommt", result.count(2), "mal vor")
result.remove(2)

print(zahlen)
print(result)

array = np.array(result)
np.savetxt("test.csv", array)
```
