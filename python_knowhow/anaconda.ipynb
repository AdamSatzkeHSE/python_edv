## Die Anaconda Distribution

Anaconda ist eine Komplettumgebung für Python, besonders beliebt für Data Science.


## Kernpunkte:

- Python + viele Pakete vorinstalliert
  - (numpy, scipy, matplotlib)
- Umgebungsmanagement mit conda (wie pip mit Extra Features)
- Jupyter Notebook integriert (interaktives Arbeiten mit Code + Text)
- Paketverwaltung einfacher als pip allein
- Besonders für Data Science / ML optimiert

## Wichtige Bestandteile

- conda: Environment und Paketmanager
- Anaconda Navigator: GUI
- Jupyter Notebook / Lab: Interaktive Notebooks

## Anaconda mit VS-Code integrieren

1) Anaconda installieren
   1) Download von anaconda.com
   2) Standard Installation reicht
   3) Nach Installation: Terminal neu starten
   4) test mit conda --version

2. Neues Conda-Environment erstellen (empfohlen)

```bash 
conda create -n myenv python=3.11
```

Aktivieren:

```bash 
conda activate myenv```
   


3. VS-Code vorbereiten

In VS-Code installieren:
- Extension: Python von Microsoft
- Extension: Jupyter

4. Conda-Interpreter in VS-Code auswählen

In VS-Code:

    1) Ctrl+Shift+P
    2) Python: Select Interpreter
    3) dein myenv auswählen

Jetzt in VS-Code dein Conda-Environment


Fehlersuche:

- Anaconda Prompt: conda --version.
  - Wenn das nicht geht, Pfad kaputt. Neu installieren mit PATH Häckchen gesetzt.
  -
