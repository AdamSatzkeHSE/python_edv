# SOLID Principles

SOLID sind fünf grundlegende Design-Prinzipien für wartbaren, erweitbaren Code. Sie stammen aus der objektorientierten Programmierung,
passen aber sehr gut zu Python, wenn man sie idiomatisch anwendet.



### Single Responsability Principle *(SRP)*

```python
# S: Single Responsability Principle (SRP)
# Eine Klasse/Funktion sollte genau einen Grund zur Änderung haben.

# Gegen SRP:
class Report:
    def generate(self):
        return "Report data"

    def save_to_file(self, filename):
        with open(filename, "w") as f:
            f.write(self.generate())

# Problem: datenlogik + Persistenz vermischt (Persistenz: daten bleiben erhalten, auch nachdem ein Programm beendet wurde.)

# SRP Konform
class Report:
    def generate(self):
        return "Report Data"

class ReportWriter:
    def save(self, report: Report, filename: str):
        with open(filename, "w") as f:
            f.write(report.generate())

# Report: Fachlogik
# ReportWriter: I/O
# unabhängig änderbar.           
```

### Open Closed Principle *(OCP)*

```python
# O - Open/Closed Principle
# Offen für Erweiterung, geschlossen für Modifikation
# Open for extension, Closed for modification

# Gegen OCP:
def calculate_discount(order, price):
    if order == "regular":
        return price * .9
    elif order == "vip":
        return price * .8
    
# Bei jedem neuen order muss die Funktion geändert werden.

# OCP - Konform
class DiscountStrategy:
    def apply(self, price):
        pass

class RegularDiscount(DiscountStrategy):
    def apply(self, price):
        return price * 0.9
    
class VipDiscount(DiscountStrategy):
    def apply(self, price):
        return price * 0.8
    
def calculate(price, strategy: DiscountStrategy):
    return strategy.apply(price)

# Erweiterung: neue Klasse
# Kein bestehender Code muss verändert werden.
```

### Liskov Substitution Principle *(LSP)*

Subtypen müssen überall dort einsetzbar sein, wo der Basistyp erwartet wird.

```python
# Gegen LSP
class Rectangle:
    def set_width(self, w):
        self.width = w
    
    def set_height(self, h):
        self.height = h

    def area(self):
        return self.width * self.height
    
class Square(Rectangle):
    def set_width(self, w):
        self.width = self.height = w

    def set_height(self, h):
        self.width = self.height = h

def print_area(rect):
    rect.set_width(4)
    rect.set_height(5)
    print(rect.area()) # erwartet 20

print_area(Square()) # gibt aber 25

# LSP - Konform (kein falsches Erben)

class Shape:
    def area(self):
        raise NotImplementedError


class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h


class Square(Shape):
    def __init__(self, size):
        self.size = size

    def area(self):
        return self.size ** 2

# Merksatz: Wenn Vererbung zu Überraschungen führt, dann lieber keine Vererbung benutzen
```

### Interface Segregation Principle
Clients sollten nicht gezwungen sein, Methoden zu implementieren, die sie nicht brauchen


```python
# Gegen ISP
class Worker:
    def work(self):
        pass

    def eat(self):
        pass

class Robot(Worker):
    def eat(self):
        pass

# ISP - Konform
class Workable:
    def work(self): ...

class Eatable:
    def eat(self): ...

class Human(Workable, Eatable):
    def work(self): 
        print("Working")
        
    def eat(self): 
        print("Eating")

class Robot(Workable):
    def work(self): print("Working")
```

### Dependency Inversion Principle

Abhängigkeiten sollen von Abstraktionen abhängen, nicht von konkreten Implementierungen.

```python
# Gegen DIP

class FileLogger:
    def log(self, msg):
        print(msg)

class UserService:
    def __init__(self):
        self.logger = FileLogger()

# Problem: UserService ist fest an FileLogger gekoppelt

# DIP - Konform

class Logger:
    def log(self, msg):
        pass

class FileLogger(Logger):
    def log(self, msg):
        print(msg)

class UserService:
    # Dependency Injection.
    def __init__(self, logger: Logger):
        self.logger = logger

    def create_user(self, name):
        self.logger.log(f"User {name} created")

service = UserService(FileLogger())

# Jetzt:
# UserService kennt nur das Verhalten, nicht die konkrete Implementierung
```
