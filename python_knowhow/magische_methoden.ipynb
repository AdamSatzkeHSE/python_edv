### Python Magische Methoden - Magic Methods

```python
# Magische Methoden in Python haben folgendes Format: __name__
# Sie werden nicht primär von dir aufgerufen, sondern vom Python-Interpreter, wenn
# bestimmte Sprachkonstrukte benutzt werden.

# len(obj) -- obj.__len__()
# obj + other -- obj.__add__(other)
# for x in obj -- obj.__iter__(), __next__()

# Sie definieren, wie sich Objekte in der Sprache verhalten.
```

```python
# Kategorien

# Objekt Lebenszyklus: __init__, __del__
# Darstellung und Debugging: __repr__, __str__
# Vergleich und Identität: __eq__, __lt__
# Container und Sammlung: __len__, __getitem__, __contains__
# Iteration: __iter__, __next__
# Kontextmanager: __enter__, __exit__
# Aufrufbarkeit: __call__
# Attributzugriff: __getattr__, __getattribute__
```

| Pattern   | Wichtige magische Methoden        |
| --------- | --------------------------------- |
| Iterator  | `__iter__`, `__next__`            |
| Composite | `__iter__`, `__len__`             |
| Strategy  | `__call__`                        |
| Command   | `__call__`                        |
| Proxy     | `__getattr__`, `__getattribute__` |
| Adapter   | `__getattr__`, `__iter__`         |
| Factory   | `__init__`, `__call__`            |
| Singleton | `__new__`                         |
| Prototype | `__copy__`, `__deepcopy__`        |
| Flyweight | `__hash__`, `__eq__`              |
