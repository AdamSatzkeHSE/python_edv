## Unpacking in Python

Unpacking bedeutet: Eine strukturierte Sammlung (Tuple, List, Dict, Iterable, Objekt) wird in einzelne Namen zerlegt.

```python

t = 1, 2, 3 # packing
a, b, c = t # unpacking

# Die rechte Seite wird auseinandergezogen und positionsweise zugewiesen.
```

```python
# Sequence Unpacking (Tuple, List, String)
a, b = (1, 2)
# Fehler: a, b = (1, 2, 3)

```

```python
# Strings sind auch Iterables
x, y, z = "abc"
```

```python
# Star - Unpacking mit *
# Ein Element "schluckt den Rest"

a, *rest = [1, 2, 3, 4]

# a == 1
# rest == [2, 3, 4]
```

```python
# Auch am Ende / in der Mitte
*anfang, ende = [1, 2, 3]
erste, *mitte, letzte = [1, 2, 3, 4, 5]

# * darf nur einmal pro Unpacking vorkommen
# das *-Ziel ist immer eine Liste.
print(anfang)
print(mitte)
```

```python
# Unpacking in for-Schleifen (sehr wichtig)
paare = [(1, "a"), (2, "b"), (3, "c")]

for zahl, buchstabe in paare:
    print(zahl, buchstabe)
    print(paare)
```

```python
# Unpacking bei Funktionen (Argumente)

def add(a, b, c):
    return a + b + c

werte = [1, 2, 3]
add(*werte)

# Liste: Positionsargumente
```

```python
# Keyword-Arguments (**kwargs)

def f(a, b):
    ...

daten = {"a": 1, "b": 2}
f(**daten)

# Dict: Keyword-Arguments
```

```python
# Kombi

f(*[1], **{"b": 2})
```

```python
a, b = {"x": 1, "y": 2}

print(a == "x")
print(b == "y")

# Items entpacken
for key, value in {"x": 1, "y": 2}.items():
    ...
```

```python
# Dicts mergen
d1 = {"a": 1}
d2 = {"b": 2}

d_neu = {**d1, **d2}
```
