## Annotations in Python

Annotations sind optionale Typ-Hinweise im Code.
Sie ändern nicht, wie Python den Code ausführt - sie sind Meta-Information


```python
# Ohne Annotations
def addiere_1(a, b):
    return a + b

# Mit Annotations
def addiere_2(a: int, b: int) -> int:
    # a: int -> Erwarterter Typ von a
    # b: int -> Erwarterter Typ von b
    # -> int: Rückgabewert ist ein int
    return a + b

# Beides geht
print(addiere_1(1, 2))
print(addiere_2(2, 3))

# Der Nutzen kommt durch:
# IDEs (VS Code, Python)
# Type Checker (mypy, pyright)
# bessere Lesbarkeit und Dokumentation
```

```python
# Union (mehrere Typen möglich)
from typing import Union

def parse(x: Union[int, str]) -> int:
    return int(x)

def parse_2(x: int | str) -> int:
    return int(x)

```

```python
# Callable (Funktionen als Typ)
from typing import Callable

def rechne(f: Callable[[int, int], int]) -> int:
    return f(2, 3)
```

```python
# Generics (z.B. eigene Container)

from typing import Generic, TypeVar

T = TypeVar("T")

class Box(Generic[T]):
    def __init__(self, value: T):
        self.value = value
```

## Was macht __future__?

"__future__" erlaubt dir, neue Sprachfeatures früher zu nutzen, bevor sie Standard werden.
Dieses Modul soll immer als erstes importiert werden.

```python
# Problem ohne __future__

# from __future__ import annotations

class Node:
    def __init__(self, next: Node | None):
        self.next = next
# Fehler: NameError: name 'Node' is not defined
# Python speichert Typen als Strings
# Keine Vorwärts Referenz probleme.


```

Nutze Annotations immer, wenn:

- du größere Projekte hast

- du mit anderen arbeitest

- du VS Code / PyCharm nutzt

- du sauberen, wartbaren Code willst

Nutze typing, wenn:

- du Container (list, dict) präzisieren willst

- None erlaubt ist

- Callables, Generics, komplexe APIs hast

Nutze from __future__ import annotations, wenn:

- Python < 3.11

- du moderne Typ-Syntax willst

- Klassen sich selbst referenzieren

- du saubere Typen ohne Strings willst
