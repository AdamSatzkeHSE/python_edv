# Listen und Tuples

```python
meine_liste = [True, 23, 9.5]

print(meine_liste)

for element in meine_liste:
    print(element)

# Indexing - Zuweisung
x = meine_liste[2]
print(x)

meine_liste[2] = 10
print(meine_liste)
```

```python
# Tupel

mein_tuple = (True, 23, 9.5)

print(mein_tuple)

for element in mein_tuple:
    print(element)

x = mein_tuple[2]
print(x)

# mein_tuple[2] = 10

```

```python
# Listen - Methoden

meine_liste.append(11)
meine_liste.append(False)

meine_liste.extend([2, 3, 4])
print(meine_liste)

meine_liste.pop()
meine_liste.remove(11)

print(meine_liste)

```

```python
index = mein_tuple.index(23)
treffer = mein_tuple.count(23)
print("Die 23 kommt", treffer, "mal vor")
print("Die 23 liegt an der", index + 1, "ten Stelle")

print(mein_tuple)
```
